```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Total no of schools - Total of all the unique schools defined by School ID.
total_schools = len(school_data['School ID'])

#Total of all the students defined by Student ID.
total_students = len(student_data['Student ID'])

#Total budget - Sum of budget column
total_budget = sum(school_data['budget'])

#Average Math Score
average_math_score = school_data_complete["math_score"].mean()

#Average reading score
average_reading_score = school_data_complete['reading_score'].mean()

#Overall passing rate - Overall Average score
average_score = (average_math_score + average_reading_score)/2

#Calculate the % of students with a 70% or greater score in maths and reading
def getStudentLengthWith70(sub_score):
    '''
        Calculate the % of students with a passing math score of 70 or greater
        Calculate the % of students with a passing reading score of 70 or greater
    '''
    temp_df = school_data_complete[school_data_complete[sub_score] >= 70]
    return (len(temp_df['Student ID'])/total_students)*100

passing_math_score = getStudentLengthWith70('math_score')
passing_reading_score = getStudentLengthWith70('reading_score')
    
#Create a summary dataframe to hold the above results
district_dicts = {'Total Schools':[total_schools], 'Total Students':[total_students], 'Total Budget':[total_budget],
                  'Average Math Score':[average_math_score], 'Average Reading Score':[average_reading_score],
                  '%Passing Math':passing_math_score, '%Passing Reading':passing_reading_score,
                  'Overall Passing Rate':average_score}
district_df = pd.DataFrame(district_dicts)
district_df.style.format({
    "Total Students":"{:,}","Total Budget":"${:,}",
    "Average Math Score":"{:.6f}", "Average Reading Score":"{:.6f}", "%Passing Math":"{:.6f}", 
    "%Passing Reading":"{:.6f}", "Overall Passing Rate":"{:.6f}"
})
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
def getSchoolSummary(schoolName):
    '''
        Returns a tuple of the following metrics:
        School Type
        Total Students
        Total School Budget
        Per Student Budget
        Average Math Score
        Average Reading Score
        % Passing Math
        % Passing Reading
        Overall Passing Rate (Average of the above two)        
    '''
    school_df = school_data_complete[school_data_complete["school_name"]== schoolName]
    df_type = school_df['type'].unique()[0]
    total_students = len(school_df['Student ID'])
    total_budget = school_df['budget'].unique()[0]
    per_student_budget = total_budget / total_students
    average_math_score = school_df['math_score'].sum()/total_students
    average_reading_score = school_df['reading_score'].sum()/total_students

    #Calculate the % of students with a 70% or greater score in maths and reading
    def getStudentLengthWith70(sub_score):
        '''
            Calculate the % of students with a passing math score of 70 or greater
            Calculate the % of students with a passing reading score of 70 or greater
        '''
        temp_df = school_df[school_df[sub_score] >= 70]
        return (len(temp_df['Student ID'])/total_students)*100

    passing_math_score = getStudentLengthWith70('math_score')
    passing_reading_score = getStudentLengthWith70('reading_score')

    #Overall passing rate
    passing_rate = (passing_math_score + passing_reading_score)/2
    
    return (schoolName, df_type, total_students, total_budget, per_student_budget, average_math_score, average_reading_score, 
            passing_math_score, passing_reading_score, passing_rate)

#Get an Array of the Schools from the School DataFrame
school_arr = school_data_complete['school_name'].unique()

#Get a list of all the school metrices
list_school_summary = [getSchoolSummary(schoolName) for schoolName in school_arr]

#Create an Overview table that summerizes the key metrices of the schools
school_summary_df = pd.DataFrame(data=list_school_summary, columns=['School Name','School Type','Total Students',
                                                                   'Total School Budget','Per Student Budget',
                                                                   'Average Math Score','Average Reading Score',
                                                                   '%Passing Math','%Passing Reading',
                                                                   '%Overall Passing Rate'])
#Zip all the Metrices and create a list
# school_tupules = list(zip(*list_school_summary))

#Refer the frame by the School Name
school_summary_df = school_summary_df.set_index('School Name')
school_summary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
def byGrade(grade, scoreType):
    '''
        This method returns Average Math and Read Scores of all the grades by school.
    '''
    school_data_grade = school_data_complete[(school_data_complete["grade"] == grade)].groupby("school_name")
    return(school_data_grade[scoreType].mean())

grade_math_dict = {}
grade_read_dict = {}
grade_arr = school_data_complete['grade'].unique()

for grade in grade_arr:
    grade_math_dict[grade] = list(byGrade(grade, 'math_score'))
    grade_read_dict[grade] = list(byGrade(grade, 'reading_score'))

math_df = pd.DataFrame(data=grade_math_dict, index=sorted(school_arr))
math_df = math_df[['9th', '10th', '11th', '12th']]
math_df
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
reading_df = pd.DataFrame(data=grade_read_dict, index=sorted(school_arr))
reading_df = reading_df[['9th', '10th', '11th', '12th']]
reading_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

#Include Spending Scores data into School Summary DataFrame
school_summary_df['Spending Scores'] = pd.cut(school_summary_df['Per Student Budget'], spending_bins, 
                                             labels=group_names)
#Group the Summary data by Spending Scores
spending_gby = school_summary_df.groupby(by='Spending Scores')

#Create a Spending Dictionary to include all the Summary Metrices.
spending_dict = {
    'Average Math Score':spending_gby['Average Math Score'].mean(),
    'Average Reading Score':spending_gby['Average Reading Score'].mean(),
    '%Passing Math':spending_gby['%Passing Math'].mean(),
    '%Passing Reading':spending_gby['%Passing Reading'].mean(),
    '%Overall Passing Rate':spending_gby['%Overall Passing Rate'].mean()    
}

#Build the Spending DataFrame
scores_school_spending_df = pd.DataFrame(data=spending_dict)
scores_school_spending_df
```
